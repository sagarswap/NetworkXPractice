```python
import matplotlib.pyplot as plt
import numpy as np
import networkx as nx
import requests
import scipy as scipy
```

```python
url='https://people.rit.edu/nxmsma/courses/data/maayan-vidal'
r=requests.get(url)
open('maayan-vidal', 'wb').write(r.content)
G=nx.read_edgelist("maayan-vidal", delimiter="\t")
nodeDegree=dict(G.degree)
freq={}
for k, v in nodeDegree.items():
    if v in freq:
        freq[v]=freq.get(v)+1
print(nodeDegree)
```
